```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fraud = pd.read_csv("Fraud_check.csv")
```

```python
colnames = list(Fraud.columns)
```

```python
predictors = colnames[:4]
```

```python
target = colnames[4]
```

```python
# np.random.uniform(start,stop,size) will generate array of real numbers with size = size

Fraud['is_train'] = np.random.uniform(0, 1, len(Fraud))<= 0.75
Fraud['is_train']
train,test = Fraud[Fraud['is_train'] == True],Fraud[Fraud['is_train']==False]
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(Fraud,test_size = 0.2)
```

```python
from sklearn.tree import  DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = 'entropy')
model.fit(train[predictors],train[target])
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
train['Undergrad'] = number.fit_transform(train['Undergrad'].astype('str'))
test['Undergrad']=number.fit_transform(test['Undergrad'].astype('str'))
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
train['Urban'] = number.fit_transform(train['Urban'].astype('str'))
test['Urban']=number.fit_transform(test['Urban'].astype('str'))
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
train['Marital.Status'] = number.fit_transform(train['Marital.Status'].astype('str'))
test['Marital.Status']=number.fit_transform(test['Marital.Status'].astype('str'))
```

```python
train.head()
```

```python
preds = model.predict(test[predictors])
pd.Series(preds).value_counts()
pd.crosstab(test[target],preds)
```

```python
# Accuracy = train

np.mean(train.Taxable_Income == model.predict(train[predictors]))
```

```python
# Accuracy = Test

np.mean(preds==test.Taxable_Income)
```
