```python
# preparing model considering given variables 

import statsmodels.formula.api as smf # for regression model
```

```python
# Preparing model     

ml1 = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=toyota).fit() # regression model
```

```python
# Getting coefficients of variables

ml1.params
```

```python
# Summary

ml1.summary()
```

```python
# preparing model based only on Age

ml_ag=smf.ols('Price~Age_08_04',data = toyota).fit()  
ml_ag.summary()
```

```python
# preparing model based only on KM

ml_km=smf.ols('Price~KM',data = toyota).fit()  
ml_km.summary()
```

```python
# preparing model based only on HP

ml_hp=smf.ols('Price~HP',data = toyota).fit()  
ml_hp.summary()
```

```python
# preparing model based only on cc

ml_cc=smf.ols('Price~cc',data = toyota).fit()  
ml_cc.summary()
```

```python
# preparing model based only on Doors

ml_dr=smf.ols('Price~Doors',data = toyota).fit()  
ml_dr.summary()
```

```python
# preparing model based only on Gears

ml_gr=smf.ols('Price~Gears',data = toyota).fit()  
ml_gr.summary()
```

```python
# preparing model based only on Quarterly_Tax

ml_qt=smf.ols('Price~Quarterly_Tax',data = toyota).fit()  
ml_qt.summary()
```

```python
# preparing model based only on Weight

ml_wt=smf.ols('Price~Weight',data = toyota).fit()  
ml_wt.summary()
```

```python
# preparing model based on cc,Doors,Gears,Quarterly Tax

ml_cdgq=smf.ols('Price~cc+Doors+Gears+Quarterly_Tax',data = toyota).fit()  
ml_cdgq.summary()
```
